```python
import matplotlib.pyplot as plt
from PIL import Image

# Load 4 images
images = [
    Image.open("sketch.jpg"),
    Image.open("image.png"),
    Image.open("lllyasviel.jpg"),
    Image.open("fine_tuned.jpg"),
]

captions = [
    "Control Image",
    "Stable Diffusion 1.5",
    "SD 1.5 + lllyasviel/sd-controlnet-canny",
    "SD 1.5 + ControlNet Trained on LSAA"
]

# Create 2x2 subplot
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # slightly bigger for clarity

for ax, img, caption in zip(axes.flatten(), images, captions):
    ax.imshow(img)
    ax.set_title(caption, fontsize=10)
    ax.axis("off")  # Hide axes

plt.tight_layout()

# Add a bottom caption for the whole figure
fig.text(0.5, 0.05, "Prompt: a side of a building in Vienna", ha='center', fontsize=12)

plt.show()
```
